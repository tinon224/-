```python
from statistics import mean


df_zone = df[['MSZoning','Neighborhood','SalePrice']].groupby(['MSZoning','Neighborhood']).mean()
mszone_list = []
nbhd_list = []
saleprice_list = []
for mszone in df_train['MSZoning'].drop_duplicates():
    df_1 = df_train[df_train['MSZoning']==mszone]
    for nbhd in df_1['Neighborhood'].drop_duplicates():
        df_2 = df_1[df_1['Neighborhood']==nbhd]
        mszone_list.append(mszone)
        nbhd_list.append(nbhd)
        saleprice_list.append(mean(list(df_2['SalePrice'])))
data = {'MSZoning':mszone_list,'Neighborhood':nbhd_list,'SalePrice_Mean':saleprice_list}
df_new = pd.DataFrame(data)
df_new = pd.merge(df_new,df_train[['MSZoning','SalePrice']].groupby(['MSZoning']).mean(),how='inner',on=['MSZoning'])
df_new = pd.merge(df_new,df_train[['Neighborhood','SalePrice']].groupby(['Neighborhood']).mean(),how='inner',on=['Neighborhood'])
df_new.columns = ['MSZoning','Neighborhood','SalePrice_Mean','SalePrice_Mean_MSZ','SalePrice_Mean_nbhd']
df_new

```
